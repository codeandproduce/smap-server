```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Init points
points = [i*10 for i in np.random.rand(80,2)]
print(points)
# Given area
area = [
    [2,8], # top left
    [8,7], # top right
    [9,4], # bottom right
    [3,2] # bottom left
]
```

```python
plt.scatter(*zip(*points))
plt.scatter(*zip(*area))


def connectpoints(x,y,p1,p2):
    if p2 == len(x):
        p2 = 0
        
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    plt.plot([x1,x2],[y1,y2],'k-')

    
for i in range(0,len(area)):
    connectpoints([x[0] for x in area], [y[1] for y in area],i, i+1)
    
    
```

## So which ones are inside the box?

### Algorithm 1: Inequalities

```python
# pick a point to work with to test:

# top line (left top to right top)
p1,p2 = area[0], area[1]
slope_12 = (p2[1] - p1[1])/(p2[0] - p1[0]) # rise over run
y_int_12 = (p2[1]-(slope_12*p2[0]))

# bottom line (left bottom to left bottom)
p1,p2 = area[3], area[2]
slope_43 = (p2[1] - p1[1])/(p2[0] - p1[0]) # rise over run
y_int_43 = (p2[1]-(slope_43*p2[0]))

# right line (right top to right bottom)
p1,p2 = area[1], area[2]
slope_23 = (p2[1] - p1[1])/(p2[0] - p1[0]) # rise over run
y_int_23 = (p2[1]-(slope_23*p2[0]))

# left line (left top to left bottom)
p1,p2 = area[0], area[3]
slope_14 = (p2[1] - p1[1])/(p2[0] - p1[0]) # rise over run
y_int_14 = (p2[1]-(slope_14*p2[0]))






# Show lines

arr12 = []
arr43 = []
arr23 = []
arr14 = []

for i in range(0,13):
    arr12.append((i*slope_12) + y_int_12)
    arr43.append((i*slope_43) + y_int_43)

for i in range(6,12):
    arr23.append((i*slope_23) + y_int_23)
    
for i in range(1,5):
    arr14.append((i*slope_14) + y_int_14)

plt.plot(range(0,13), arr12)
plt.plot(range(0,13), arr43)
plt.plot(range(6,12), arr23)
plt.plot(range(1,5), arr14)
    
plt.show()
```

```python
def check_if_inside_bounds(point):
    x, y = point[0], point[1]
    # top line
    if y < (x*slope_12)+y_int_12:
        # bottom line
        if y > (x*slope_43)+y_int_43:
            # right line
            if y < (x*slope_23) + y_int_23:
                # left line
                if y > (x*slope_14) + y_int_14:
                    return True
    return False
```

```python
inside_bounds_arr = []

for i in points:
    if check_if_inside_bounds(i):
        inside_bounds_arr.append(i)

print("points inside bounds:")
print(inside_bounds_arr)

plt.scatter(*zip(*points))
plt.scatter(*zip(*area))

plt.plot(range(0,13), arr12)
plt.plot(range(0,13), arr43)
plt.plot(range(6,12), arr23)
plt.plot(range(1,5), arr14)



plt.scatter(*zip(*inside_bounds_arr), c='r')

plt.show()
```

```python
def right(h,x):
    return h + x + np.sqrt(h**2 + x**2)

def not_right(h,x):
    k = 2*(h**2)
    l = (x**2)/2
    j = ((2*h)-1)*x
    return k+l-j

arr = np.random.rand(1000,2)*10
print()


count = 0
for i in range(0, len(arr)):
    arr_this = arr[i]
    if arr_this[0] < arr_this[1]:
        temp = arr_this[0]
        arr_this[0] = arr_this[1]
        arr_this[1] = temp
        
    print("array:")
    print(arr_this)
    print("right: ")
    r = right(arr_this[0], arr_this[1])
    print(r)
    
    print("not_right:")
    nr = not_right(arr_this[0], arr_this[1])
    print(nr)
    
    if nr<r:
        arr_this=arr_this+1
    else:
        print(True)

    print()

print("count::")
print(count)
```
